# Momentum

```python
prices = pd.DataFrame()
return_rate = pd.DataFrame()

for i in sp500_list:
    print (i)
    yf.pdr_override()
    data = pdr.get_data_yahoo(i, start="2000-01-31", end="2021-12-31")
    closing_data = pd.DataFrame(data['Close'])
    closing_data = closing_data.reset_index()
    closing_data['Date'] = closing_data['Date'].apply(lambda x: x.strftime('%Y-%m'))
    closing_data.set_index('Date', inplace=True)
    #Group df by column and get the first value in each group                   
    grouped_df = closing_data.groupby("Date").first()

    #Reset indices to match format
    first_values = grouped_df.reset_index()
    first_values['Date']=pd.to_datetime(first_values['Date'], format='%Y-%m-%d')
    first_values=first_values.set_index('Date')
    first_values[i] = first_values['Close']
    first_values=first_values.drop(['Close'],axis=1)
    prices=pd.concat((prices,first_values),axis=1)
    first_values[i] = first_values[i].pct_change(1)
    return_rate=pd.concat((return_rate,first_values),axis=1)
```

```python
return_rate.to_csv('return_rate.csv')
prices.to_csv('prices.csv')
```
